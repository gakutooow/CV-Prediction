# trainデータとtestデータに分ける

```python
df = pd.read_csv("../data/feature/new_monthly/feature_df_201706.csv",index_col=0)
df_train = pd.DataFrame(columns=df.columns)

year_month_list = [
                   "201711",
                   "201712",
                   "201801",
                   "201802",
                   "201803",
                   "201804",
                  ]

for year_month in year_month_list:
    df = pd.read_csv("../data/feature/new_monthly/feature_df_{}.csv".format(year_month),index_col=0)

    df_train = pd.concat([df_train,df])
    
year_month = "201805"
df_test = pd.read_csv("../data/feature/new_monthly/feature_df_{}.csv".format(year_month),index_col=0)


#閲覧数１を削除
df_train = df_train[df_train["num_of_pages"] > 1]
df_test = df_test[df_test["num_of_pages"] > 1]

```

```python
#sessionidが被ってしまう➡️月を跨いだアクセスの可能性あり➡︎削除

train_use_session_list = []
for k,v in dict(Counter(df_train["session_id"])).items():
    if v == 1:
        train_use_session_list.append(k)

test_use_session_list = []
for k,v in dict(Counter(df_test["session_id"])).items():
    if v == 1:
        test_use_session_list.append(k)        

df_train = df_train[df_train["session_id"].isin(train_use_session_list)]
df_test = df_test[df_test["session_id"].isin(test_use_session_list)]



df_train_cv = df_train[df_train["cv_flg"]==1]
df_train_non_cv = df_train[df_train["cv_flg"]==0]

df_test_cv = df_test[df_test["cv_flg"]==1]
df_test_non_cv = df_test[df_test["cv_flg"]==0]


print(df_train_cv.shape)
print(df_train_non_cv.shape) ##ここはアンダーサンプリング必要

print(df_test_cv.shape)
print(df_test_non_cv.shape)
```

# ランダムにアンダーサンプリング

```python
df_train_non_cv["random_index"] = np.arange(len(df_train_non_cv))

# sample_index = random.sample(list(df_train_non_cv["random_index"]), len(df_train_cv))
sample_index = random.sample(list(df_train_non_cv["random_index"]), len(df_train_cv))
df_train_non_cv = df_train_non_cv[df_train_non_cv["random_index"].isin(sample_index)]

df_train_non_cv = df_train_non_cv.drop("random_index",axis=1)

df_train = pd.concat([df_train_cv,df_train_non_cv])
df_test = pd.concat([df_test_cv,df_test_non_cv])


print(df_train.shape)
print(df_test.shape)
```

```python
df_train_dummy = pd.DataFrame()
df_train_dummy["session_id"] = df_train_non_cv["session_id"]
df_train_dummy["cv_flg"] = df_train_non_cv["cv_flg"]
df_train_dummy["num_of_pages"] = df_train_non_cv["num_of_pages"]
df_train_dummy["num_of_move_block"] = df_train_non_cv["num_of_move_block"]
df_train_dummy["num_of_kind_block"] = df_train_non_cv["num_of_kind_block"]

df_train_dummy = pd.merge(df_train_dummy,dummy_last_block,on="session_id")
print(df_train_dummy.shape)
df_train_dummy = pd.merge(df_train_dummy,dummy_channelGrouping,on="session_id")
print(df_train_dummy.shape)
df_train_dummy = pd.merge(df_train_dummy,dummy_device_deviceCategory,on="session_id")
print(df_train_dummy.shape)
df_train_dummy = pd.merge(df_train_dummy,dummy_geoNetwork_region,on="session_id")
print(df_train_dummy.shape)
df_train_dummy = pd.merge(df_train_dummy,dummy_geoNetwork_country,on="session_id")
print(df_train_dummy.shape)


#トラフィックソースはクラスタリングに使用しない
# df_train_dummy = pd.merge(df_train_dummy,dummy_trafficSource_source,on="session_id")

df_train_dummy = pd.merge(df_train_dummy,dummy_weekday,on="session_id")
print(df_train_dummy.shape)

df_train_dummy = df_train_dummy.drop(["session_id","cv_flg"],axis=1)
```

# クラスター数の決定

```python
distortions = []

for i  in range(1,2):
    print(i)
    km = KMeans(n_clusters=i,
                init='k-means++',     # k-means++法によりクラスタ中心を選択
                n_init=10,
                max_iter=300,
                random_state=0)
    km.fit(df_train_dummy.values)                         # クラスタリングの計算を実行
    distortions.append(km.inertia_)   # km.fitするとkm.inertia_が得られる

plt.plot(range(1,8),distortions,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances ')
plt.grid(True)
plt.show()
```
